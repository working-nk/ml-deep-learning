# Loading Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer.feature_names)
```

```python
print(cancer.DESCR)
```

```python
cancer['data'].shape
```

```python
# converting into dataframe
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns= cancer['feature_names'].tolist() + ['target'])
```

```python
df.head()
```

# Visualizing Data

```python
df.describe()
```

```python
sns.pairplot(df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness'])
```

```python
sns.countplot(df['target'] )
```

```python
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df)
```

```python
plt.figure(figsize = (20, 10))
sns.heatmap(df.corr(), annot = True)
```

# Model Training

```python
y = df['target']
x = df.drop(['target'], axis = 1)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = SVC()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=5)
```

```python
model.fit(x_train, y_train)
```

# Model Evaluation

```python
model.score(x_test, y_test)
```

```python
y_predicted = model.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_predicted)
```

```python
sns.heatmap(cm)
```

```python
sns.heatmap(cm, annot = True)
```

### We should Normalize our data, use different parameters - C Parameter, Gamma Parameter

#### Normalization

```python
min_train = x_train.min()
```

```python
range_train = (x_train-min_train).max()
```

```python
x_train_scaled = (x_train - min_train)/range_train
```

```python
sns.scatterplot(x_train['mean area'], x_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = x_test.min()
range_test = (x_test-min_test).max()
x_test_scaled = (x_test - min_test)/range_test
```

```python
model.fit(x_train_scaled, y_train)
```

```python
model.score(x_test_scaled, y_test)
```

```python
y_predicted_new = model.predict(x_test_scaled)

cm = confusion_matrix(y_test, y_predicted_new)

sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_predicted_new))
```

#### Parameter Improvision

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), param_grid, refit = True, verbose = 4)
```

```python
grid.fit(x_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
grid_predict = grid.predict(x_test_scaled)
```

```python
grid.score(x_test_scaled, y_test)
```

```python
cm = confusion_matrix(y_test, grid_predict)

sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, grid_predict))
```
